```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
csv_1 = '/Volumes/USB_DRIVE/NEW/2025-06-10_A.csv'
csv_2 = '/Volumes/USB_DRIVE/NEW/2025-06-03_A.csv'
csv_all = '2HDM_A_test.csv'
df_1 = pd.read_csv(csv_1)
df_2 = pd.read_csv(csv_2)
df_all = pd.read_csv(csv_all)
plt.scatter(df_1['m'], df_1['tanb'], c='red', label='2025-06-10', alpha=0.5)
plt.scatter(df_2['m'], df_2['tanb'], c='blue', label='2025-06-03', alpha=0.5)
plt.scatter(df_all['m'], df_all['tanb'], c='green', label='All Data', alpha=0.1)
plt.xlabel('m')
plt.ylabel('tanβ')
plt.title('Scatter Plot of tanβ vs m')
plt.legend()
plt.savefig('tanb_vs_m_scatter_plot.png')
```
