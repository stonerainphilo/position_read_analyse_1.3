```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def quadratic_fit(csv_file):
    # 读取CSV文件
    data = pd.read_csv(csv_file)
    
    # 假设CSV文件有两列，分别是 x 和 y
    x = data.iloc[:, 0].values
    y = data.iloc[:, 1].values
    
    # 使用numpy进行二次拟合
    coefficients = np.polyfit(x, y, 2)  # 二次多项式拟合
    quadratic_function = np.poly1d(coefficients)
    
    # 打印拟合的二次函数
    print("拟合的二次函数为:")
    print(quadratic_function)
    
    # 可视化原始数据和拟合曲线
    plt.scatter(x, y, color='blue', label='original')  # 原始数据点
    x_fit = np.linspace(min(x), 5, 500)  # 生成拟合曲线的x值
    y_fit = quadratic_function(x_fit)  # 计算拟合曲线的y值
    plt.plot(x_fit, y_fit, color='red', label='fit')  # 拟合曲线
    plt.xlabel('x')
    plt.ylabel('y')
    plt.xlim(0, 5)
    plt.legend()
    plt.title('fit')
    plt.show()
    
    return coefficients

quadratic_fit('f0_A.csv')
```
