```python
csv_file = 'theta_tau.csv'
df = pd.read_csv(csv_file)
x = df['mH']
y = df['tau']
plt.plot(x, y, 'o', label='data')
plt.xlabel('mH')
plt.ylabel('tau')
plt.title('tau vs mH')
plt.xscale('log')
plt.yscale('log')
plt.legend()
plt.grid()
# plt.savefig('tau_vs_mH.png')
plt.show()
```

```python
def Gamma_run(csv_file, theta):
    df = pd.read_csv(csv_file)
    theta0 = 1e-6
    df[f'tau_{theta}'] = df['tau'] * theta0**2 / theta**2
    df.to_csv(f'tau_theta_{theta}.csv', index=False)
    return df
    
    
def LSD_tau_diff(csv_file, theta):
    theta0 = 1e-6
    df = pd.read_csv(csv_file)
    df[f'tau_{theta}'] = df['ltime'] * theta0**2 / theta**2
    print(df[f'tau_{theta}'][:5])
    return df
```
