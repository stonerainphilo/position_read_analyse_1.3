```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.append("/media/ubuntu/6156e08b-fdb1-4cde-964e-431f74a6078e/Program/PRA/Github/position_read_analyse_1.3/ALL_IN_ONE")
import Pythia8.functions_for_calculation as fcal
```

```python
final_csv = '/media/ubuntu/SRPPS/Results/2HDM_H/C_fast.csv'
final_df = pd.read_csv(final_csv)
cross_section = 3800000
# final_df = final_df[final_df['CODEX-b_acceptance']>0]
df_CODEX = final_df[final_df['CODEX-b_acceptance'] * 300 * cross_section * fcal.calcu_Br_B_H_2HDM_I_simple(final_df['m'], final_df['tanb'])>3]
df_FASER2 = final_df[final_df['FASER2_acceptance'] * 300 * cross_section * fcal.calcu_Br_B_H_2HDM_I_simple(final_df['m'], final_df['tanb'])>3]
df_MATHUSLA = final_df[final_df['MATHUSLA_acceptance'] * 300 * cross_section * fcal.calcu_Br_B_H_2HDM_I_simple(final_df['m'], final_df['tanb'])>3]
plt.figure(figsize=(10, 6))
# plt.fill_between(final_df['m'], final_df['tanb'], color='skyblue', alpha=0.5)
plt.scatter(df_CODEX['m'], df_CODEX['tanb'], marker='o', color='blue', label='CODEX-b Acceptance', alpha=0.3)
plt.scatter(df_FASER2['m'], df_FASER2['tanb'], marker='s', color='green', label='FASER2 Acceptance', alpha=0.3)
plt.scatter(df_MATHUSLA['m'], df_MATHUSLA['tanb'], marker='^', color='red', label='MATHUSLA Acceptance', alpha=0.3)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Mass (m)')
plt.ylabel('tan')
plt.title('Exclusion Region based on Significance > 3')
plt.legend()
plt.grid(True, which="both", ls="--", linewidth=0.5)
plt.show()
```
